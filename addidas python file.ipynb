```python
##loading the dataset
df = pd.read_csv("Adida_data.xlsx - Data Sales Adidas (2) (2).csv")
df.head()
```

```python
## checking for hidden figures
for column in df:
  print(repr(column))
```

```python
## checking the column names
print(df.columns.tolist())
```

```python
## droping the irrelevant column
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df
```

```python
## checking for duplicates
df.duplicated().sum()
```

```python
## checking for data type
df.dtypes
```

```python
## converting the date to the proper format
df["Invoice Date"] = pd.to_datetime(df["Invoice Date"]).dt.strftime('%Y-%m-%d')
df
```

```python
#removing symbols and converting to float
columns_to_clean = ['Price per Unit', 'Total Sales', 'Units Sold', 'Operating Profit']
for column in columns_to_clean:
    df[column] = df[column].astype(str)
    df[column] = df[column].replace('[\$,]',"" , regex=True)
    df[column] = df[column].str.replace(',', '',  regex=False)
    df[column] = df[column].str.strip()
    df[column] = df[column].astype(float)
```

```python
df
```

```python
#converting to float

df['Operating Margin']= df['Operating Margin'].replace('[%]',"",regex=True).astype(float)/100
```

```python
##exporting the cleaned date
df.to_csv('cleaned_adidas_data.csv', index=False)
```
